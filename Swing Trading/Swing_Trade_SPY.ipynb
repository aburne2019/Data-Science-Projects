```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
spy = yf.Ticker("SPY").history(period='max') # pulls SPY index fund and queries all history of the data for the index fund
```

```python
del spy['Dividends']
del spy['Stock Splits']
del spy['Capital Gains']
```

```python
spy.tail()
```

```python
spy_1month = spy.tail(20)
spy_1month.plot.line(y='Close', use_index=True)
```

```python
x_values = spy_1month.index
y_values = spy_1month['Close']
```

```python
x = [x_values[0], x_values[18]]
y = [y_values[0], y_values[18]]

y_15 = [y_values[0], y_values[0]*(1+0.0066)]
```

```python
plt.plot(x_values, y_values, 'bo', linestyle="-")
plt.plot(x, y, 'ro', linestyle='--')
plt.plot(x, y_15, 'go', linestyle='--')
plt.xticks(rotation=90)
plt.title('1 Month Projection')
plt.legend(['Close', 'Baseline', '8% Projection'])
plt.show()
```

```python
spy_3month = spy.tail(90)
spy_3month.plot.line(y='Close', use_index=True)

x_values = spy_3month.index
y_values = spy_3month['Close']

x = [x_values[0], x_values[88]]
y = [y_values[0], y_values[88]]

y_15 = [y_values[0], y_values[0]*(1+0.0066*3)]

plt.plot(x_values, y_values, 'b', linestyle="-")
plt.plot(x, y, 'ro', linestyle='--')
plt.plot(x, y_15, 'go', linestyle='--')
plt.xticks(rotation=90)
plt.title('3 Month Projection')
plt.legend(['Close', 'Baseline', '8% Projection'])
plt.show()
```

```python
spy_year = spy.tail(365)
spy_year.plot.line(y='Close', use_index=True)

x_values = spy_year.index
y_values = spy_year['Close']

x = [x_values[0], x_values[363]]
y = [y_values[0], y_values[363]]

y_15 = [y_values[0], y_values[0]*(1+0.0066*12)]

plt.plot(x_values, y_values, 'b', linestyle="-")
plt.plot(x, y, 'ro', linestyle='--')
plt.plot(x, y_15, 'go', linestyle='--')
plt.xticks(rotation=90)
plt.title('1 Year Projection')
plt.legend(['Close', 'Baseline', '8% Projection'])
plt.show()
```

```python
spy_week = spy.tail(5)

x_values = spy_week.index
y_values = spy_week['Close']

x = [x_values[0], x_values[3]]
y = [y_values[0], y_values[3]]

y_15 = [y_values[0], y_values[0]*(1+.08/52)]

plt.plot(x_values, y_values, 'b', linestyle="-")
plt.plot(x, y, 'ro', linestyle='--')
plt.plot(x, y_15, 'go', linestyle='--')
plt.xticks(rotation=90)
plt.title('1 Week Projection')
plt.legend(['Close', 'Baseline', '8% Projection'])
plt.show()
```

```python
def projection(days):
    spy_proj = spy.tail(days)
    #spy_proj.plot.line(y='Close', use_index=True)

    x_values = spy_proj.index
    y_values = spy_proj['Close']

    x = [x_values[0], x_values[days-2]]
    y = [y_values[0], y_values[days-2]]

    y_15 = [y_values[0], y_values[0]*(1+(.08/365)*days)]

    plt.plot(x_values, y_values, 'b', linestyle="-")
    plt.plot(x, y, 'ro', linestyle='--')
    plt.plot(x, y_15, 'go', linestyle='--')
    plt.xticks(rotation=90)
    plt.title(f'{days} Days Projection')
    plt.legend(['Close', 'Baseline', '8% Projection'])
    plt.show()
    
```

```python
projection(10)
```
