```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
spy = yf.Ticker("SPY").history(period='max') # pulls SPY index fund and queries all history of the data for the index fund
```

```python
spy.tail()
```

```python
spy_1month = spy.tail(20)
spy_1month.plot.line(y='Close', use_index=True)
```

```python
x_values = spy_1month.index
y_values = spy_1month['Close']
```

```python
x = [x_values[0], x_values[18]]
y = [y_values[0], y_values[18]]

y_15 = [y_values[0], y_values[0]*(1+0.0066)]
```

```python
plt.plot(x_values, y_values, 'bo', linestyle="-")
plt.plot(x, y, 'ro', linestyle='--')
plt.plot(x, y_15, 'go', linestyle='--')
plt.xticks(rotation=90)
plt.title('1 Month Projection')
plt.show()
```

```python
spy_3month = spy.tail(90)
spy_3month.plot.line(y='Close', use_index=True)

x_values = spy_3month.index
y_values = spy_3month['Close']

x = [x_values[0], x_values[88]]
y = [y_values[0], y_values[88]]

y_15 = [y_values[0], y_values[0]*(1+0.0066*3)]

plt.plot(x_values, y_values, 'b', linestyle="-")
plt.plot(x, y, 'ro', linestyle='--')
plt.plot(x, y_15, 'go', linestyle='--')
plt.xticks(rotation=90)
plt.title('3 Month Projection')
plt.show()
```
